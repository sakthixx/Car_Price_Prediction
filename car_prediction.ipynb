```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data collection and processing

```python
#load the dataset from the dataframe
car_dataset = pd.read_csv('/kaggle/input/vehicle-dataset-from-cardekho/car data.csv')
```

```python
#displaying first 5 rows of the dataset
car_dataset.head()
```

```python
#checking total number of rows and columns
car_dataset.shape
```

```python
#getting some information about the dataset
car_dataset.info()
```

```python
#check the number of missing values
car_dataset.isnull().sum()
```

```python
# As these are categorical data, check the distribution of these categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

Encoding the categorical data

```python
# Encode the data, as ML model do not understand text
#for "Fuel_Type" column
car_dataset.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2}}, inplace=True)

#for "Seller_Type" column
car_dataset.replace({'Seller_Type':{'Dealer':0, 'Individual':1}}, inplace=True)

#for "Transmission" column
car_dataset.replace({'Transmission':{'Manual':0, 'Automatic':1}}, inplace=True)
```

```python
car_dataset.head()
```

Splitting the data and target

```python
x = car_dataset.drop(['Car_Name', 'Selling_Price'], axis=1)
y = car_dataset['Selling_Price']
```

```python
print(x)
```

```python
print(y)
```

splitting training data and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=2)
```

Model training

> Linear Regression

```python
# Loading linear regression model
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(x_train, y_train)
```

Model evaluation

```python
# Prediciton of training data
training_data_prediction = lin_reg_model.predict(x_train)
```

```python
# Compare the predicted values using R squared error
error_score = metrics.r2_score(y_train, training_data_prediction)
print("R-Squared Error: ", error_score)
```

Visualize the actual Prices and pedicted prices

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual_Price")
plt.ylabel("Predicted_Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
# Prediciton of training data
test_data_prediction = lin_reg_model.predict(x_test)
```

```python
# Compare the predicted values using R squared error
error_score = metrics.r2_score(y_test, test_data_prediction)
print("R-Squared Error: ", error_score)
```
